### Eliminar los duplicados de la columna de género

```python
movie_temp['genre'].value_counts()
```

```python
movie_temp['genre']=[name.lower() for name in movie_temp['genre']]
movie_temp['genre'].value_counts()
```

```python
movie_temp['genre']=movie_temp['genre'].replace('infantil','children')
movie_temp['genre']=movie_temp['genre'].replace('infantil ','children')
movie_temp['genre']=movie_temp['genre'].replace('acción','action')
movie_temp['genre']=movie_temp['genre'].replace('aventura','adventure')
movie_temp['genre']=movie_temp['genre'].replace('comedia','comedy')
movie_temp['genre']=movie_temp['genre'].replace('suspenso','thrillers')
movie_temp['genre']=movie_temp['genre'].replace('ficción','science fiction')
movie_temp['genre']=movie_temp['genre'].replace('misterio','mistery')
movie_temp['genre']=movie_temp['genre'].replace('crimen','crime')
movie_temp['genre']=movie_temp['genre'].replace('fantasía','fantasy')
movie_temp['genre']=movie_temp['genre'].replace('biografía','biopics')
movie_temp['genre']=movie_temp['genre'].replace('biográfica','biopics')
movie_temp['genre']=movie_temp['genre'].replace('ciencia ficción','science fiction')
movie_temp['genre']=movie_temp['genre'].replace('drama juvenil','drama')
movie_temp['genre']=movie_temp['genre'].replace('romance ','romance')
```
