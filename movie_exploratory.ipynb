```python
#libraries 

import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
movies = pd.read_csv('movie_database_clean.csv')
movies = pd.read_csv('movie_database_clean.csv',index_col='title')
```

```python
movies.head()
```

```python
#working with missing values 
movies.isnull().sum() #dice cuántos datos son nulos, todos deberían tener 0
    
```

### Sacar el promedo de las columnas de person_1 a person_20 y eliminarlas

```python
movies.iloc[:,1:22]
```

```python
movies.iloc[:,1:21].mean()
```

```python
movies.iloc[:,1:21].mean(axis=1)
```

### Cleaning columns

```python
movies['fan_rating']=movies.iloc[:,1:21].mean(axis=1)
movies.head()
```

```python
dropped_columns=movies.iloc[:,1:21].columns
dropped_columns
```

```python
movies.drop(columns=dropped_columns,inplace=True)
movies.head()
```

### Rellenar las casillas vacías por los promedios 

```python
movies.isnull().sum()
```

```python
movies.dropna()
```

```python
movies[movies['personal_rating'].isnull()]
```

```python
movies[movies['budget'].isnull()]
```

```python
movies[movies['box_office'].isnull()]
```

```python
personal_mean=movies['personal_rating'].mean()
budget_mean=movies['budget'].mean()
box_office_mean=movies['box_office'].mean()
personal_mean,budget_mean,box_office_mean
```

```python
movies['personal_rating'].fillna(personal_mean,inplace=True)
movies['budget'].fillna(budget_mean,inplace=True)
movies['box_office'].fillna(box_office_mean,inplace=True)
movies.isnull().sum()
```

```python
duration_mean=movies['duration'].mean()
episode_mean=movies['episode'].mean()
seasons_mean=movies['seasons'].mean()
duration_mean,episode_mean,seasons_mean
```

### Realizar lo mismo con las filas que corresponden a las series 

```python
movies['duration'].fillna(duration_mean,inplace=True)
movies['episode'].fillna(episode_mean,inplace=True)
movies['seasons'].fillna(seasons_mean,inplace=True)
movies.isnull().sum()
```

### Estadística descriptiva

```python
movies.describe()
```

```python
movies['year']=movies['year'].astype(str)
movies.describe()
```

```python
movies_numeric = movies.select_dtypes(include='number')
movies_numeric.describe()
```

### Detectar y eliminar outliers

```python
fig, axes = plt.subplots(6, 1,figsize=(5,12))
fig.tight_layout(pad=3)
sns.boxplot(x=movies_numeric['personal_rating'], ax=axes[0]);
sns.boxplot(x=movies_numeric['fan_rating'], ax=axes[1]);
sns.boxplot(x=movies_numeric['duration'], ax=axes[2]);
sns.boxplot(x=movies_numeric['budget'], ax=axes[3]);
sns.boxplot(x=movies_numeric['box_office'], ax=axes[4]);
sns.boxplot(x=movies_numeric['imdb'], ax=axes[5]);
```

```python
Q1 = movies_numeric.quantile(0.25)
Q3 = movies_numeric.quantile(0.75)
IQR = Q3 - Q1
Q1
```

```python
movies_numeric[((movies_numeric < (Q1 - 1.5 * IQR)) |(movies_numeric > (Q3 + 1.5 * IQR))).any(axis=1)]
```

```python
movies_duration = movies_numeric['duration']
movies_duration[((movies_duration < (Q1['duration'] - 1.5 * IQR['duration'])) |(movies_duration > (Q3['duration'] + 1.5 * IQR['duration'])))]
```

```python
movies_duration.drop(['The amazing Spiderman','Piratas del caribe: El cofre de la muerte'])
```

```python
movies.drop(['The amazing Spiderman','Piratas del caribe: El cofre de la muerte'],inplace=True)
movies.shape
```

```python
movies_numeric = movies.select_dtypes(include='number')
fig, axes = plt.subplots(5, 1,figsize=(5,10))
fig.tight_layout(pad=3)
sns.boxplot(x=movies_numeric['personal_rating'], ax=axes[0]);
sns.boxplot(x=movies_numeric['fan_rating'], ax=axes[1]);
sns.boxplot(x=movies_numeric['duration'], ax=axes[2]);
sns.boxplot(x=movies_numeric['budget'], ax=axes[3]);
sns.boxplot(x=movies_numeric['box_office'], ax=axes[4]);
```

### Eliminar outliers the box office 

```python
movies_box_office = movies_numeric['box_office']
movies_box_office[((movies_box_office < (Q1['box_office'] - 1.5 * IQR['box_office'])) |(movies_box_office > (Q3['box_office'] + 1.5 * IQR['box_office'])))]
```

```python
movies_box_office.drop(['Joker','Super Mario Bros','Batman El Caballero de la Noche Asciende','Batman El Caballero de la Noche','The Avengers'])
```

```python
movies.drop(['Batman El Caballero de la Noche','The Avengers'],inplace=True)
movies.shape
```
